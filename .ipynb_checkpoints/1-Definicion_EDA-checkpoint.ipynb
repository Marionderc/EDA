```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='User Rating', hue='Genre', data=df_libros)
plt.title('Tendencias de Calificación de Usuarios a lo Largo de los Años')
plt.show()
#las personas que leen ficción compran mas en amazon (auqnue haya un pico de bajada)
#dejan mas nota los que leen ficción
```

```python
libros_repetidos = df_libros.groupby('Name').size().reset_index(name='Frecuencia')
libros_repetidos = libros_repetidos[libros_repetidos['Frecuencia'] > 1]
libros_repetidos
#vamos a ver que libros se repiten a lo largo de los años
```

```python
libros_repetidos.describe()
```

```python
libros_repetidos = libros_repetidos[libros_repetidos['Frecuencia'] > 4 ]
libros_repetidos
```

```python
libros_repetidos = pd.merge(libros_repetidos, df_libros[['Name', 'Genre']], on='Name', how='left')
libros_repetidos
#utilizo la función merge para ver los valores comunes
```

```python
libros_repetidos_sin_duplicados = libros_repetidos.drop_duplicates(subset='Name')
print(libros_repetidos_sin_duplicados[['Name', 'Genre']])
```

```python
libros_repetidos_sin_duplicados = pd.merge(libros_repetidos_sin_duplicados, df_libros[['Name', 'Year']], on='Name', how='left')
print(libros_repetidos_sin_duplicados[['Name', 'Genre', 'Year']])
```

```python
libros_sin_duplicados = libros_repetidos_sin_duplicados.drop_duplicates(subset=['Name', 'Genre', 'Year'])
libros_sin_duplicados
```

```python
libros_repetidos = df_libros.groupby('Name').size().reset_index(name='Frecuencia')
libros_repetidos = libros_repetidos.sort_values(by='Frecuencia', ascending=False)
libros_repetidos.head(10)
```

```python
autores_repetidos = df_libros.groupby('Author').size().reset_index(name='Frecuencia')
autores_repetidos = autores_repetidos.sort_values(by='Frecuencia', ascending=False)
top_10_autores = autores_repetidos.head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='Frecuencia', y='Author', data=top_10_autores, hue='Author', palette='viridis', dodge=False, legend=False)
plt.title('Top 10 Autores que Más se Repiten a lo Largo de los Años')
plt.xlabel('Frecuencia')
plt.ylabel('Autor')
plt.show()
```

```python
#con género

autores_repetidos = df_libros.groupby('Author').size().reset_index(name='Frecuencia')
autores_y_generos = df_libros.groupby(['Author', 'Genre']).size().reset_index(name='Count')
autores_y_generos = autores_y_generos.sort_values(by='Count', ascending=False).drop_duplicates('Author')
autores_repetidos = pd.merge(autores_repetidos, autores_y_generos[['Author', 'Genre']], on='Author', how='left')
autores_repetidos = autores_repetidos.sort_values(by='Frecuencia', ascending=False)
top_10_autores = autores_repetidos.head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='Frecuencia', y='Author', data=top_10_autores, hue='Genre', palette='viridis', dodge=False)
plt.title('Top 10 Autores que Más se Repiten a lo Largo de los Años con Género')
plt.xlabel('Frecuencia')
plt.ylabel('Autor')
plt.show()
```
