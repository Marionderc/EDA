```python
# Tendencias a lo largo del tiempo, es la cantidad total de libros vendidos por año

total_libros_vendidos = df_libros.groupby('Year')['Reviews'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Reviews', data=total_libros_vendidos, marker='o')
plt.title('Total de reseñas de libros más vendidos por año')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Suponiendo que df_libros es tu DataFrame y tienes una columna 'Author' que representa el autor
# Crear un DataFrame con los autores y sus ventas totales
autores_mas_vendidos = df_libros['Author'].value_counts().reset_index(name='Total Ventas')

# Ordenar los autores por sus ventas totales de forma descendente
autores_mas_vendidos = autores_mas_vendidos.sort_values(by='Total Ventas', ascending=False)

# Visualizar los 10 autores más vendidos
plt.figure(figsize=(15, 8))
sns.barplot(x='Total Ventas', y='index', data=autores_mas_vendidos.head(10), palette='viridis')
plt.title('Los 10 Autores Más Vendidos')
plt.xlabel('Número de Ventas')
plt.ylabel('Autor')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Suponiendo que df_libros es tu DataFrame y tienes una columna 'Author' que representa el autor
# Crear un DataFrame con los autores y sus ventas totales
autores_mas_vendidos = df_libros['Author'].value_counts().reset_index(name='Total Ventas')

# Ordenar los autores por sus ventas totales de forma descendente
autores_mas_vendidos = autores_mas_vendidos.sort_values(by='Total Ventas', ascending=False)

# Visualizar los 10 autores más vendidos
plt.figure(figsize=(15, 8))
sns.barplot(x='Total Ventas', y='Author', data=autores_mas_vendidos.head(10), hue='Author', palette='viridis', legend=False)
plt.title('Los 10 Autores Más Vendidos')
plt.xlabel('Número de Ventas')
plt.ylabel('Autor')
plt.show()

```

```python

# Crear un DataFrame con los autores, los años y sus ventas totales
autores_por_anio = df_libros.groupby(['Author', 'Year']).size().reset_index(name='Total Ventas')

# Filtrar los 5 principales autores por año
top_autores_por_anio = autores_por_anio.groupby('Year').apply(lambda x: x.nlargest(5, 'Total Ventas')).reset_index(drop=True)

# Visualizar los 5 principales autores por año
plt.figure(figsize=(15, 8))
sns.lineplot(x='Year', y='Total Ventas', hue='Author', data=top_autores_por_anio, palette='viridis', marker='o')
plt.title('Evolución de los 5 Autores Más Vendidos por Año')
plt.xlabel('Año')
plt.ylabel('Número de Ventas')
plt.legend(title='Autor', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
